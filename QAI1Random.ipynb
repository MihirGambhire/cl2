```python
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister, transpile
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram, circuit_drawer
```

```python
circuit = QuantumCircuit(16,16)
circuit.h(range(16))
circuit.measure(range(16),range(16))
circuit.draw()
```

```python
backend = AerSimulator()
output = transpile(circuit,backend)
output = backend.run(output,shots=1).result()
output.get_counts()
```

```python
"""
Explanation:

QuantumCircuit → used to build the quantum circuit (collection of qubits + operations).

QuantumRegister & ClassicalRegister → optional, used to define qubits and classical bits.

AerSimulator → simulates a real quantum device (since we’re not using actual IBM hardware).

plot_histogram, circuit_drawer → used for visualization of results and circuit structure.

transpile → optimizes the circuit for a particular backend (here, AerSimulator).   

"""
```
